```python
import pandas as pd
import re
pd.set_option('display.max_rows', 500)

df = pd.read_csv('u_cmdb_ci_appl_bus_app.csv', encoding='utf-8')



df['u_implementation'] = df['u_implementation'].str.replace(r'<[^<>]*>', '', regex=True)
df = df.applymap(str)
df = df[df.u_implementation != 'nan']
df = df[df.u_implementation != 'Other']
df = df[df.u_implementation != 'Unknown']
df = df[df.u_implementation != 'Not Applicable']
df = df[df.u_implementation != 'Azure CF2']
df = df[df.u_implementation != 'Hybrid']
df = df[df.u_implementation != 'Multi-Cloud']
plot = df['u_implementation'].value_counts()
ax = plot.plot(kind="pie", autopct='%1.1f%%', title="Hosted at", labeldistance=None )
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plot
```
